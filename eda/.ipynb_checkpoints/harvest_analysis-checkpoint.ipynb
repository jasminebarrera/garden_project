### <p style="color:#87CEEB">Harvest Analysis</p>

Questions to explore:
1. Which crop produced the most overall? The least?
2. Of the most productive crops, which varieties produced the most?
3. How did harvest frequency change over time?
4. How did harvest productivity change over time?

<br>
Q1. Analyze Crop Productivity

```python
# Most productive crops by total harvest weight
harvest_weight_sum = harvesting_df.groupby(['crop'])['weight'].sum().reset_index().rename(columns= {'weight':'total_weight'})

top_n_crops = harvest_weight_sum.sort_values(by=['total_weight'], ascending=False)[0:5]

sns.catplot(data=top_n_crops, x="crop", y="total_weight", kind='bar')
plt.xlabel('Crop')
plt.ylabel('Weight (g)')
plt.title('Top Crops By Total Harvest Weight (g)')
plt.xticks(rotation=50)
plt.show()
```

```python
# Least productive crops by total harvest weight
bottom_n_crops = harvest_weight_sum.sort_values(by=['total_weight'], ascending=True)[0:5]

sns.catplot(data=bottom_n_crops, x="crop", y="total_weight", kind='bar')
plt.xlabel('Crop')
plt.ylabel('Weight (g)')
plt.title('Bottom Crops By Total Harvest Weight (g)')
plt.xticks(rotation=50)
plt.show()
```

<br>
Q2. Analyze Crop Productivity by Variety

```python
# Most productive crop varieties by total harvest weight
top_n_crops_list = top_n_crops['crop'].unique()

crop_dict = dict.fromkeys(top_n_crops_list)

for crop in top_n_crops_list:
    crop_list = harvesting_df.loc[harvesting_df['crop'] == crop]
    grouped_crop_var = crop_list.groupby(['variety'])['weight'].sum().reset_index().rename(columns= {'weight':'total_weight'})
    crop_dict[crop] = grouped_crop_var.sort_values(by=['total_weight'], ascending=False)[0:5]
    print(crop, crop_dict[crop], '\n')
```

<br>
Q3. Analyze Number of Harvests Over Time

```python
# Recall that there were 114 unique 'date' values in the dataset
harvest_count_by_date = harvesting_df.groupby(['date'])['weight'].count().reset_index().rename(columns= {'weight':'harvest_count'})

sns.lineplot(data=harvest_count_by_date, x='date', y='harvest_count')
plt.xlabel('Date')
plt.ylabel('Harvest Count')
plt.title('Harvest Count Over Time')
plt.xticks(rotation=50)

plt.show()
```

<br>
Q4. Analyze Crop Productivity Over Time

```python
# Recall that there were 114 unique 'date' values in the dataset
harvest_weight_by_date = harvesting_df.groupby(['date'])['weight'].sum().reset_index().rename(columns= {'weight':'total_weight'})
# print(harvest_weight_by_date)

sns.lineplot(data=harvest_weight_by_date, x="date", y="total_weight")
plt.xlabel('Date')
plt.ylabel('Total Weight (g)')
plt.title('Total Harvest Weight (g) Over Time')
plt.xticks(rotation=50)

plt.show()
```

### <p style="color:#87CEEB"> Analysis Summary</p>

The top five <ins>most productive crops</ins> in terms of total harvest weight were **tomatoes, pumpkins, zucchini, squash, cucumbers**. The top five <ins>least productive crops</ins> were **chives, asparagus, cilantro, apple, kohlrabi**.

Among the most productive crops, some varieties produced more than others.
The amish paste, better boy, grape, and old german tomato varieties were the most productive, producing a total of 72028 grams (159 lbs) of tomatoes.
The most productive pumpkin varieties were New England sugar and cinderella's carriage, producing a total of 35259 grams (78 lbs) of pumpkins.
The romanesco zucchini variety produced a total of 45227 grams (100 lbs).
The blue, waltham butternut, red kuri, and delicata varieties were the most productive, producing a total of 44917 grams (99 lbs).
The pickling cucumber variety produced a total of 19781 grams (44 lbs).

According to the data, daily harvests steadily increased between early June through early August, with some high harvest days between early September and early October. The garden increased production between August and early September. There were some harvest peaks during and after September that coincided with pumpkin, squash, and zucchini harvests. These findings coincide with peak summer harvest season and cool, fall temperatures following the summer garden season.

With these findings in mind, future garden plans can take into account which crops and varieties produce the most as well as what to expect in terms of harvesting frequency and production during busy summer months.
